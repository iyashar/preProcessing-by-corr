```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# https://github.com/hannarud/r-plotting/blob/master/Train_UWu5bXk.csv
data = pd.read_csv('Train_UWu5bXk.csv')
data.head()
```

```python
data['Item_Weight'].fillna(data['Item_Weight'].median(), inplace=True)
data['Outlet_Size'].fillna(data['Outlet_Size'].mode()[0], inplace=True)
```

```python
numeric_data = data[['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Outlet_Establishment_Year', 'Item_Outlet_Sales']]
var = numeric_data.var()
numeric_columns = numeric_data.columns
keep_columns = []
for i in range(0, len(var)):
    if var[i] >= 10:
        keep_columns.append(numeric_columns[i])
print(keep_columns)
```

```python
numeric_data = data[['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Outlet_Establishment_Year', 'Item_Outlet_Sales']]
corr = numeric_data.corr()
#np.eye:: یک آرایه 2 بعدی با نوع مشخص شده بر می گرداند که قطر اصلی آن خالی یا صفر است.
#.mask():: مواردی را بر می گرداند که داخل پردانتز درخواست شده و باقی را خالی می گذارد یا با مقداری که بدهیم پر می کند
m = ~(corr.mask(np.eye(len(corr), dtype=bool)).abs() > 0.5).any()
raw = corr.loc[m, m]
raw
```
